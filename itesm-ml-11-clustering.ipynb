```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster import vq
```

```python
datos = np.loadtxt('datos2D3Clusters.txt')
print(datos[:4, :])
```

```python
plt.scatter(datos[:,0], datos[:,1])
plt.show()
```

```python
# Calcular los centroides con K-Means
v = []
for k in range(1, 4):
    centroides, varianza = vq.kmeans(datos, k)
    v.append(varianza)
    print(f'K:{k}\n Centroides:\n{centroides} \nVarianza: {varianza}')
plt.plot(v)
plt.show()    
```

```python
# Obtener los grupos
grupo, distancia = vq.vq(datos, centroides)
print(f'Grupo:\n{grupo}\nDistancia:\n{distancia}')
```

```python
# Separando los grupos finales
grupo0 = datos[grupo == 0]
grupo1 = datos[grupo == 1]
grupo2 = datos[grupo == 2]
print(f'Grupo 01: {len(grupo0)}\nGrupo 02: {len(grupo1)}\nGrupo 03: {len(grupo2)}')
```

```python
# Graficando los centroides y datos
plt.scatter(grupo0[:,0], grupo0[:,1], linewidths=0.5, c = 'g')
plt.scatter(grupo1[:,0], grupo1[:,1], linewidths=0.5, c = 'y')
plt.scatter(grupo2[:,0], grupo2[:,1], linewidths=0.5, c = 'r')

plt.plot(centroides[:, 0], centroides[:, 1], 'kx', markersize = 15)
plt.show()
```

```python
# Pronosticando datos nuevos
datos_pronostico = [[2,2], [6,3], [8,8]]
grupo_pronostico, distancia_pronostico = vq.vq(datos_pronostico, centroides)
print(grupo_pronostico)
```
