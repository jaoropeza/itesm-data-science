```python
from tensorflow.keras import models
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos = np.loadtxt('datos2.txt')
np.random.shuffle(datos)
x = datos[:,:2]
y = datos[:,-1]
```

```python
print('Tamaño Datos:', datos.shape)
print('Tamaño X, Y: ', x.shape, y.shape)
m = len(y)
limite = int(0.8 * m)
xTrain, xTest = x[:limite, :], x[limite:, :]
yTrain, yTest = y[:limite], y[limite:]

print('xTrain, xTest: ', xTrain.shape, xTest.shape)
print('yTrain, yTest: ', yTrain.shape, yTest.shape)
```

```python
# La arquitectura de la red
red = models.Sequential()
layer1 = layers.Dense(20, activation='relu', input_shape=(2, ))
layer2 = layers.Dense(3, activation='softmax')
red.add(layer1)
red.add(layer2)
```

```python
# Compilación de la red
red.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Preparando las etiquetas (salidas)
from tensorflow.keras.utils import to_categorical
print(yTrain)
yTrain = to_categorical(yTrain)
print(yTrain)
yTest = to_categorical(yTest)
```

```python
# Entrenar la red
red.fit(xTrain, yTrain, epochs=25, batch_size=128)
```

```python
# Verificando el modelo con el conjunto de datos Test
test_loss, test_accuracy = red.evaluate(xTest, yTest)
```

```python
# Probando algunos valores extras
pre = np.array([[5,2], [7,6], [3,6], [4,7], [7,0], [9,6]])
predicciones = red.predict(pre)
print(predicciones)
res = np.argmax(predicciones, axis=1)
print(res)
```
