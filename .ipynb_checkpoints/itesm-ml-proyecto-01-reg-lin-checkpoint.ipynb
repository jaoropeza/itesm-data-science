```python
# Carga de la información con la que trabajaremos.
datos = np.loadtxt('datos1.txt')
datos = np.transpose(datos)
```

```python
class Practica:
    def __init__(self, learning_rate, epochs, datos, p):
        # Inicializamos hiper-parámetros.
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.p = p
        
        # Particionamos datos.
        self.set_data(datos)

        # Definimos los tensores
        self.X = tf.constant(self.xTrain, dtype=tf.float32)
        self.Y = tf.constant(self.yTrain, dtype=tf.float32)
        self.T = tf.Variable([0.]*self.p, dtype=tf.float32)
        
        # Definimos "learning_rate" al optimizador
        self.train_op = tf.keras.optimizers.SGD(self.learning_rate)
        
    def set_data(self, datos):
        # Particionamos datos.
        x = datos[0]
        y = datos[1]
        self.m = len(y)

        self.xTrain, self.xTest = x[:160], x[-40:]
        self.yTrain, self.yTest = y[:160], y[-40:]

    def modelo(self, T, X):
        # Calculamos el grado del polinomio.
        n = len(T.numpy())
        m = 0
        for i in range(n):
            m = m + T[i]*(X**i)
        return m

    def costo(self, T, X, Y):
        m = len(Y.numpy())
        return tf.reduce_sum(tf.square(self.modelo(T, X) - Y))/(2*m)

    def costo_minimizar(self):
        return tf.reduce_sum(tf.square(self.modelo(self.T, self.X) - self.Y))/(2*self.m)

    def fit(self):
        self.hist_error = []
        for epoch in range(self.epochs):
            self.error = self.costo(self.T, self.X, self.Y).numpy()
            self.hist_error.append(self.error)
            self.train_op.minimize(self.costo_minimizar, var_list=[self.T])
        print(f'T: {self.T.numpy()}, Error:{self.error}')

    def predict(self):
        XT = tf.constant(self.xTest, dtype = tf.float32)
        YT = tf.constant(self.yTest, dtype = tf.float32)
        print(f'Error: {self.error}')
        print(f'Costo: {self.costo(self.T, XT, YT).numpy()}')
        print(f'Dif. Error y Costo: {self.error-self.costo(self.T, XT, YT).numpy()}')

    def plot_error(self):
        plt.plot(self.hist_error)
        plt.title(f'T{self.p}')
        plt.show()
        
    def plot_data(self):
        # Visualización de las información para generar el modelo.
        plt.title('Datos')
        plt.scatter(self.xTrain, self.yTrain, c='b')
        plt.scatter(self.xTest, self.yTest, c='r')
        plt.show()
        
    def all(self):
        self.fit()
        self.predict()
        self.plot_error()

p1 = Practica(0.005, 50, datos, 3)
p1.plot_data()
p1.all()
p2 = Practica(0.000005, 50, datos, 4)
p2.all()
p3 = Practica(0.0000005, 20, datos, 5)
p3.all()
```
